## 1. 取出今天所有的新聞

```python
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime
import time

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

today_date=datetime.today().date()

while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    #抓取目前網頁上，最後一則新聞的時間：.find_all('h3')[-1]
    last_news_date_txt = soup.find(class_="part_list_2").find_all('h3')[-1].find(class_="date").text
    last_news_date = datetime.strptime(last_news_date_txt, "%Y/%m/%d %H:%M").date()
    time.sleep(2)
    if today_date != last_news_date:
        print("now the last news datetime is: ",last_news_date_txt, "for checking")
        print("end of today's news")
        break

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    news_date = datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M").date()
    if news_date == today_date:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import timedelta

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

current_timestamp=datetime.now()
#兩個小時前的時間設定
stop_timestamp = current_timestamp + timedelta(hours=-2)

while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    last_news_date_txt = soup.find(class_="part_list_2").find_all('h3')[-1].find(class_="date").text
    last_news_date = datetime.strptime(last_news_date_txt, "%Y/%m/%d %H:%M")
    time.sleep(2)
    #設定停止條件，當最後一條新聞的時間，低於現在的兩個小時前
    if stop_timestamp >= last_news_date:
        print("now the last news datetime is: ",last_news_date_txt, "for checking")
        print("end of news scrolling")
        break

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    news_date = datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if news_date >= stop_timestamp:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select

#開啟瀏覽器
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

#三天前
three_date_ago = (current_timestamp + timedelta(days=-3))

time.sleep(10)

#點選三天前的按鈕
selectYear = Select(browser.find_element_by_id("selY"))
selectYear.select_by_value(datetime.strftime(three_date_ago,'%Y'))
selectMonth = Select(browser.find_element_by_id("selM"))
selectMonth.select_by_value(datetime.strftime(three_date_ago,'%#m'))
selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value(datetime.strftime(three_date_ago,'%#d'))

browser.find_element_by_id('button').click()

#設定下午3點到5點的時間
start_datetime = three_date_ago.replace(hour=15,minute=0,second=0,microsecond=0)
end_datetime = three_date_ago.replace(hour=17,minute=0,second=0,microsecond=0)

#開始循環，直到時間出現低於3點時停止
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    last_news_date_txt = soup.find(class_="part_list_2").find_all('h3')[-1].find(class_="date").text
    last_news_date = datetime.strptime(last_news_date_txt, "%Y/%m/%d %H:%M")
    time.sleep(2)
    if start_datetime >= last_news_date:
        print("now the last news datetime is: ",last_news_date_txt, "for checking")
        print("end of news scrolling")
        break
#show
for d in soup.find(class_="part_list_2").find_all('h3'):
    news_date = datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if news_date >= start_datetime and news_date <= end_datetime:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
