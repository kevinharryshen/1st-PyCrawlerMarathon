```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d = {}
for tr in table.find_all('tr'):
    if tr.find("td",attrs={'class':'no-alt'}):
        tag = tr.find("td",attrs={'class':'no-alt'}).string
        record = []
        for td in tr.find_all("td"):
            record.append(td.text)
        d.setdefault(tag,{})
        d[tag][record[-3]] = record[-2]
    else:
        if tr.find("td"):
            record = []
            for td in tr.find_all("td"):
                record.append(td.text)
            d[tag][record[0]] = record[1]
print(d)
```
